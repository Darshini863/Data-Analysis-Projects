```python
df_filter = df.sort_values(['revenue'], ascending = False).head(500)
df_filter
```

```python
df_central = df_filter.loc[:,['budget',"revenue","runtime"]]
df_central
```

```python
import statistics as st

Mean = []
Median = []
Mode = []

for col in df_central:
    Mean.append(df_central[col].mean())
    Median.append(df_central[col].median())
    Mode.append(st.mode(df_central[col]))
    
row_heads = ["mean", "Median", "Mode"]
col_name = ["Budget", "Revenue", "Runtime"]

d_data = pd.DataFrame([Mean, Median, Mode], columns = col_name)
d_data
```

```python
d_data.insert(0,"measures",row_heads)
d_data
```

```python
fig,ax=plt.subplots(1,3,figsize=(15,7))
sns.boxplot(y=df_central['budget'],ax=ax[0],color='g')
sns.boxplot(y=df_central['revenue'],ax=ax[1],color='r')
sns.boxplot(y=df_central['runtime'],ax=ax[2],color='b')
plt.show()
```

```python
df_central
```

```python
# outlier analysis for Budget
mean = int(df_central.budget.mean())
mean
x = df_central[df_central['budget'] > 280000000].index
x
```

```python
for index in x:
    df_central.loc[index, 'budget'] = mean
```

```python
# outlier analysis for Revenue
mean = int(df_central.revenue.mean())
mean
x = df_central[df_central['revenue'] > mean].index
x
```

```python
for index in x:
    df_central.loc[index, 'revenue'] = mean
```

```python
# outlier Analysis for Runtime
mean = int(df_central.runtime.mean())
mean
x = df_central[df_central['runtime'] > 178].index
x
```

```python
for index in x:
    df_central.loc[index, 'runtime'] = mean
```

```python
fig,ax=plt.subplots(1,3,figsize=(15,7))
sns.boxplot(y=df_central['budget'],ax = ax[0], color='g')
sns.boxplot(y = df_central['revenue'], ax = ax[1], color = 'y')
sns.boxplot(y = df_central['runtime'], ax = ax[2], color = 'b')
```
